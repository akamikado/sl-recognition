```python
path='C:\\Users\\vc\\Downloads\\dataset\\archive\\Gesture Image Data\\'
image_size=224
images=[]
labels=[]
for i in sign_types:
    data_path=path+str(i)
    filenames=[i for i in os.listdir(data_path)]
    for f in filenames:
        image = tf.io.read_file(data_path+'/'+f)
        image = tf.image.decode_jpeg(image)
        image=tf.image.resize(image,[image_size,image_size])
        images.append(image)
        labels.append(i)
labels
```

```python
def identity_block(x,f,filters):
    
    F1,F2,F3=filters
    
    x_shortcut=x

    x=layers.Conv2D(filters=F1, kernel_size=(1,1), strides=(1,1), padding='valid')(x)
    x=layers.BatchNormalization(axis=3)(x)
    x=layers.Activation('relu')(x)

    x=layers.Conv2D(filters=F2, kernel_size=(f,f), strides=(1,1), padding='same')(x)
    x=layers.BatchNormalization(axis=3)(x)
    x=layers.Activation('relu')(x)

    x=layers.Conv2D(filters=F3, kernel_size=(1,1), strides=(1,1), padding='valid')(x)
    x=layers.BatchNormalization(axis=3)(x)

    x=layers.Add()([x,x_shortcut])
    x=layers.Activation('relu')(x)

    return x
```

```python
def convolution_block(x,f,filters,s=2):

    F1,F2,F3=filters

    x_shortcut=x

    x=layers.Conv2D(F1,kernel_size=(1,1),strides=(s,s))(x)
    x=layers.BatchNormalization(axis=3)(x)
    x=layers.Activation('relu')(x)

    x=layers.Conv2D(filters=F2, kernel_size=(f,f), strides=(1,1), padding='same')(x)
    x=layers.BatchNormalization(axis=3)(x)
    x=layers.Activation('relu')(x)

    x=layers.Conv2D(filters=F3, kernel_size=(1,1), strides=(1,1), padding='valid')(x)
    x=layers.BatchNormalization(axis=3)(x)

    x_shortcut=layers.Conv2D(filters=F3,kernel_size=(1,1), strides=(s,s), padding='valid')(x_shortcut)
    x_shortcut=layers.BatchNormalization(axis=3)(x_shortcut)

    x=layers.Add()([x,x_shortcut])
    x=layers.Activation('relu')(x)

    return x
```

```python
def ResNet50(input_shape=(224,224,3),classes=2):

    x_input=Input(shape=input_shape)

    x=layers.ZeroPadding2D((3,3))(x_input)

    x=layers.Conv2D(64,(7,7), strides=(2,2))(x)
    x=layers.BatchNormalization(axis=3)(x)
    x=layers.Activation('relu')(x)
    x=layers.MaxPooling2D((3,3), strides=(2,2))(x)

    x=convolution_block(x,f=3,filters=[64,64,256],s=1)
    x=identity_block(x,3,[64,64,256])
    x=identity_block(x,3,[64,64,256])

    x=convolution_block(x,f=3,filters=[128,128,256],s=2)
    x=identity_block(x,3,[128,128,256])
    x=identity_block(x,3,[128,128,256])
    x=identity_block(x,3,[128,128,256])

    x=convolution_block(x,f=3,filters=[256,256,1024],s=2)
    x=identity_block(x,3,[256,256,1024])
    x=identity_block(x,3,[256,256,1024])
    x=identity_block(x,3,[256,256,1024])
    x=identity_block(x,3,[256,256,1024])
    x=identity_block(x,3,[256,256,1024])

    x=convolution_block(x,f=3,filters=[512,512,2048],s=2)
    x=identity_block(x,3,[512,512,2048])
    x=identity_block(x,3,[512,512,2048])

    x=layers.AveragePooling2D((2,2))(x)
    x=layers.Flatten()(x)
    x=layers.Dense(classes,activation='softmax')(x)

    model=models.Model(inputs=x_input,outputs=x)
    return model
```

```python
model=ResNet50(input_shape=(224,224,3),classes=2)
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau, TensorBoard

early_stopping = EarlyStopping(monitor='val_loss', patience=10, mode='min')
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=0.001)

callbacks = [early_stopping, reduce_lr]
```

```python
model.summary()
```

```python
model.fit(train_x,train_y,epochs=3,batch_size=32, callbacks=callbacks, validation_data=(val_x,val_y))
```

```python
preds=model.evaluate(test_x,test_y)
print("loss " + str(preds[0]))
print("test accuracy= " + str(preds[1]))
```

```python
import cv2
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input, decode_predictions
from ultralytics import YOLO

# Load YOLOv8 model
yolo_model = YOLO('yolov8n.pt')  # You can choose a different model size, e.g., yolov8s.pt, yolov8m.pt, etc.

# Load ResNet50 model for classification
#resnet_model = ResNet50(weights='imagenet')

# Function to get the bounding boxes from YOLOv8
def get_bounding_boxes(image):
    results = yolo_model(image)
    boxes = []
    h, w = image.shape[:2]
    
    for result in results:
        for bbox in result.boxes:
            x_min, y_min, x_max, y_max = map(int, bbox.xyxy)
            boxes.append([x_min, y_min, x_max - x_min, y_max - y_min])
    
    return boxes

# Function to classify the objects using ResNet50
def classify_objects(image, boxes):
    classifications = []
    for box in boxes:
        x, y, w, h = box
        roi = image[y:y+h, x:x+w]
        roi = cv2.resize(roi, (224, 224))
        roi = preprocess_input(roi)
        roi = np.expand_dims(roi, axis=0)
        
        preds = ResNet50.predict(roi)
        decoded_preds = decode_predictions(preds, top=1)[0][0]
        classifications.append(decoded_preds)
    
    return classifications

# Combined function to detect and classify objects
def detect_and_classify(image):
    boxes = get_bounding_boxes(image)
    classifications = classify_objects(image, boxes)
    
    for box, classification in zip(boxes, classifications):
        x, y, w, h = box
        label = "{}: {:.2f}".format(classification[1], classification[2])
        cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.putText(image, label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    
    return image

# Example usage
image_path = "C:\\Users\\vc\\Downloads\\dataset\\archive\\Gesture Image Data\\3\\1021.jpg"
image = cv2.imread(image_path)
#for im in images:
result_image = detect_and_classify(image)

# Display the result
cv2.imshow("Result", result_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
